```python
from numpy import *
from matplotlib.pyplot import *

# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, Aer, transpile, assemble, execute

from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
from qiskit.providers.aer import QasmSimulator


# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
cr= ClassicalRegister(2,"bell_basis_measurement")
alice_qubit= QuantumRegister(1, 'alice__qubit')
bob_qubit= QuantumRegister(1, 'bob__qubit')
lq1= QuantumRegister(8, 'ancilla_qubit_alice')
lq2= QuantumRegister(8,'ancilla_qubit_bob')
                                            
repetation_code_qc= QuantumCircuit(alice_qubit,lq1,bob_qubit,lq2,cr)
                                            
repetation_code_qc.h(alice_qubit[0])
repetation_code_qc.cx(alice_qubit[0],bob_qubit[0])
                                            
repetation_code_qc.barrier()

repetation_code_qc.cx(0,[3,6])
repetation_code_qc.cx(9,[12,15])

repetation_code_qc.h([0,3,6,9,12,15])

repetation_code_qc.cx(0,[1,2])
repetation_code_qc.cx(3,[4,5])
repetation_code_qc.cx(6,[7,8])
repetation_code_qc.cx(9,[10,11])
repetation_code_qc.cx(12,[13,14])
repetation_code_qc.cx(15,[16,17])

repetation_code_qc.barrier()

repetation_code_qc.z([9])
repetation_code_qc.x([4,9])

repetation_code_qc.barrier()

repetation_code_qc.cx(0,[1,2])
repetation_code_qc.cx(3,[4,5])
repetation_code_qc.cx(6,[7,8])
repetation_code_qc.cx(9,[10,11])
repetation_code_qc.cx(12,[13,14])
repetation_code_qc.cx(15,[16,17])

repetation_code_qc.h([0,3,6,9,12,15])

repetation_code_qc.cx(0,[3,6])
repetation_code_qc.cx(9,[12,15])
repetation_code_qc.ccx(3,6,0)
repetation_code_qc.ccx(12,15,9)

repetation_code_qc.barrier()
repetation_code_qc.cx(alice_qubit[0],bob_qubit[0])
repetation_code_qc.h(alice_qubit[0])
repetation_code_qc.barrier()
repetation_code_qc.measure([alice_qubit[0], bob_qubit[0]],cr)
repetation_code_qc.draw()
```
